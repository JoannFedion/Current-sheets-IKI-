```python
from scipy.integrate import solve_ivp
```

```python
def plasma(t,z,a,b,c,d):
    x, y, n = z
    return [n*y, x*(1-n),-x*y*n/(a*b*n**(b-1)+c/2-(d**2/n**2))]
```

```python
sol = solve_ivp(plasma,[0,4],[0,0.412,1.42], args=(0.5, 5/3, 0.5, 0.1), rtol = 1e-14)
```

```python
import numpy as np
```

```python
>>> t = sol.t
>>> z = sol.y

>>> import matplotlib.pyplot as plt
>>> plt.plot(t, z.T)
plt.xlabel('z')
>>> plt.legend(['B_x', 'u_y','n'], shadow=True)
>>> plt.show()
```
